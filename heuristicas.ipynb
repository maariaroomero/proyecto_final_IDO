```python
def calcular_distancia(ciudad1, ciudad2):
    """
    Calcula la distancia euclidiana entre dos ciudades.
    ciudad1 y ciudad2 son tuplas (x, y).
    """
    return round(math.sqrt((ciudad1[0] - ciudad2[0])**2 + (ciudad1[1] - ciudad2[1])**2))

def leer_datos(archivo):
    """
    Lee el archivo con las coordenadas de las ciudades y devuelve una lista de ciudades.
    Cada ciudad es una tupla (x, y).
    """
    ciudades = []
    with open(archivo, 'r') as f:
        lineas = f.readlines()[1:]  # Ignorar el encabezado
        for linea in lineas:
            datos = linea.split()
            if len(datos) == 3:  # Asegurarse de que la línea tiene datos válidos
                x, y = float(datos[1]), float(datos[2])
                ciudades.append((x, y))
    return ciudades

def vecino_mas_cercano(ciudades):
    """
    Aplica el algoritmo de vecino más cercano para resolver el TSP.
    Devuelve el recorrido y la distancia total.
    """
    n = len(ciudades)
    visitadas = [False] * n
    recorrido = []
    distancia_total = 0

    # Iniciar en la primera ciudad
    ciudad_actual = 0
    visitadas[ciudad_actual] = True
    recorrido.append(ciudad_actual)

    for _ in range(n - 1):
        siguiente_ciudad = None
        distancia_minima = float('inf')

        # Buscar la ciudad más cercana no visitada
        for i in range(n):
            if not visitadas[i]:
                distancia = calcular_distancia(ciudades[ciudad_actual], ciudades[i])
                if distancia < distancia_minima:
                    distancia_minima = distancia
                    siguiente_ciudad = i

        # Moverse a la siguiente ciudad
        visitadas[siguiente_ciudad] = True
        recorrido.append(siguiente_ciudad)
        distancia_total += distancia_minima
        ciudad_actual = siguiente_ciudad

    # Regresar a la ciudad inicial
    distancia_total += calcular_distancia(ciudades[ciudad_actual], ciudades[recorrido[0]])
    recorrido.append(recorrido[0])

    return recorrido, distancia_total

# Archivo con datos (cambiar por la ruta del archivo proporcionado)
archivo = "Qatar.txt"
ciudades = leer_datos(archivo)

# Resolver el TSP
recorrido, distancia_total = vecino_mas_cercano(ciudades)

# Resultados
print("Recorrido:", recorrido)
print("Distancia total:", distancia_total)
```

```python

def calcular_distancia(ciudad1, ciudad2):
    """
    Calcula la distancia euclidiana entre dos ciudades.
    ciudad1 y ciudad2 son tuplas (x, y).
    """
    return round(math.sqrt((ciudad1[0] - ciudad2[0])**2 + (ciudad1[1] - ciudad2[1])**2))

def leer_datos(archivo):
    """
    Lee el archivo con las coordenadas de las ciudades y devuelve una lista de ciudades.
    Cada ciudad es una tupla (x, y).
    """
    ciudades = []
    with open(archivo, 'r') as f:
        lineas = f.readlines()[1:]  # Ignorar el encabezado
        for linea in lineas:
            datos = linea.split()
            if len(datos) == 3:  # Asegurarse de que la línea tiene datos válidos
                x, y = float(datos[1]), float(datos[2])
                ciudades.append((x, y))
    return ciudades

def vecino_mas_cercano(ciudades):
    """
    Aplica el algoritmo de vecino más cercano para resolver el TSP.
    Devuelve el recorrido y la distancia total.
    """
    n = len(ciudades)
    visitadas = [False] * n
    recorrido = []
    distancia_total = 0

    # Iniciar en la primera ciudad
    ciudad_actual = 0
    visitadas[ciudad_actual] = True
    recorrido.append(ciudad_actual)

    for _ in range(n - 1):
        siguiente_ciudad = None
        distancia_minima = float('inf')

        # Buscar la ciudad más cercana no visitada
        for i in range(n):
            if not visitadas[i]:
                distancia = calcular_distancia(ciudades[ciudad_actual], ciudades[i])
                if distancia < distancia_minima:
                    distancia_minima = distancia
                    siguiente_ciudad = i

        # Moverse a la siguiente ciudad
        visitadas[siguiente_ciudad] = True
        recorrido.append(siguiente_ciudad)
        distancia_total += distancia_minima
        ciudad_actual = siguiente_ciudad

    # Regresar a la ciudad inicial
    distancia_total += calcular_distancia(ciudades[ciudad_actual], ciudades[recorrido[0]])
    recorrido.append(recorrido[0])

    return recorrido, distancia_total

def optimizar_2opt(ciudades, recorrido):
    """
    Mejora un recorrido dado utilizando la heurística 2-opt.
    Devuelve el nuevo recorrido y la distancia total optimizada.
    """
    n = len(recorrido) - 1  # El último nodo es igual al primero en un ciclo
    mejora = True

    while mejora:
        mejora = False
        for i in range(1, n - 1):
            for j in range(i + 1, n):
                if j - i == 1:  # Evitar intercambios consecutivos
                    continue

                # Calcular el cambio de distancia si se realiza el intercambio
                d1 = calcular_distancia(ciudades[recorrido[i - 1]], ciudades[recorrido[i]])
                d2 = calcular_distancia(ciudades[recorrido[j]], ciudades[recorrido[(j + 1) % n]])
                d3 = calcular_distancia(ciudades[recorrido[i - 1]], ciudades[recorrido[j]])
                d4 = calcular_distancia(ciudades[recorrido[i]], ciudades[recorrido[(j + 1) % n]])

                if d3 + d4 < d1 + d2:
                    # Realizar el intercambio
                    recorrido[i:j + 1] = recorrido[i:j + 1][::-1]
                    mejora = True

    # Calcular la distancia total después de la optimización
    distancia_total = sum(
        calcular_distancia(ciudades[recorrido[k]], ciudades[recorrido[k + 1]]) for k in range(n)
    )
    return recorrido, distancia_total

# Archivo con datos (cambiar por la ruta del archivo proporcionado)
archivo = "Qatar.txt"
ciudades = leer_datos(archivo)

# Resolver el TSP con Vecino Más Cercano
recorrido_inicial, distancia_inicial = vecino_mas_cercano(ciudades)
print("Recorrido inicial:", recorrido_inicial)
print("Distancia inicial:", distancia_inicial)

# Optimizar con 2-opt
recorrido_opt, distancia_opt = optimizar_2opt(ciudades, recorrido_inicial)
print("Recorrido optimizado:", recorrido_opt)
print("Distancia optimizada:", distancia_opt)
```
